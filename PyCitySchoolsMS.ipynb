# PyCity Schools Analysis

- It can strongly be inferred that students attending charter schools (as opposed to district schools) achieved much higher test scores across the board, especially math. In turn, this led to a higher overall passing rate for charter schools (90.43%) which is close to double that of district schools (53.67%).

Another interesting find within the analysis of this dataset is the seemingly negative correlation between spending per student and the percent of students who are passing. As spending ranges (per student) decrease, the percent of students passing increase and vice versa. This trend can also be seen with school budget; as budget increases the percent of students passing decreases.

---

```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path

# File to Load (Remember to Change These)
school_data_to_load = Path("Resources/schools_complete.csv")
student_data_to_load = Path("Resources/students_complete.csv")

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.
school_data_complete = pd.merge(student_data, school_data, how="left", on="school_name")
school_data_complete.head()
```

## District Summary

```python
school_data_complete["school_name"].unique()
```

```python
# Calculate the total number of unique schools 
# use len function
school_count = len(school_data_complete["school_name"].unique())
school_count
```

```python
# Calculate the total number of students
student_count = len(school_data_complete["student_name"])
student_count
```

```python
school_data
```

```python
# Calculate the total budget
total_budget = sum(school_data["budget"])
total_budget
```

```python
# Calculate the average (mean) math score
average_math_score = (sum(school_data_complete["math_score"])) / (len(school_data_complete["math_score"]))
average_math_score
```

```python
# Calculate the average (mean) reading score
average_reading_score = school_data_complete["reading_score"].mean()
average_reading_score
```

```python
# use conditional to find all students who had a score of 70 or higher in math
passMathStudents = school_data_complete[school_data_complete["math_score"] >= 70]
# set variable for number of students who passed math
numPassMath = len(passMathStudents)
numPassMath
```

```python
# Use the following to calculate the percentage of students who passed math (math scores greather than or equal to 70)
percentagePassedMath = (numPassMath / student_count) * 100
percentagePassedMath
```

```python
# Calculate the percentage of students who passed reading (hint: look at how the math percentage was calculated)
passReadStudents = school_data_complete[school_data_complete["reading_score"] >= 70]
# set variable for number of students who passed reading
numPassRead = len(passReadStudents)
numPassRead
```

```python
# Calculate the percentage of students who passed reading (hint: look at how the math percentage was calculated)
percentagePassedRead = (numPassRead / student_count) * 100
percentagePassedRead
```

```python
# Use the following to calculate the percentage of students that passed math and reading
passingStudents = school_data_complete[
    (school_data_complete["math_score"] >= 70) & (school_data_complete["reading_score"] >= 70)]
#passingStudents.head()
numPassingStudents = len(passingStudents)
numPassingStudents
```

```python
# Use the following to calculate the percentage of students that passed math and reading
overall_passing_rate = (numPassingStudents / student_count) * 100
overall_passing_rate
```

```python
# Create a high-level snapshot of the district's key metrics in a DataFrame
districtSummary = pd.DataFrame(
    {
        "Total Schools": [school_count],
        "Total Students": [student_count],
        "Total Budget": [total_budget],
        "Average Math Score": [average_math_score],
        "Average Reading Score": [average_reading_score],
        "Percentage Passing Math": [percentagePassedMath],
        "Percentage Passing Reading": [percentagePassedRead],
        "Percentage Overall Passing": [overall_passing_rate]
    }
)

# Formatting
districtSummary["Total Students"] = districtSummary["Total Students"].map("{:,}".format)
districtSummary["Total Budget"] = districtSummary["Total Budget"].map("${:,.2f}".format)
districtSummary["Average Math Score"] = districtSummary["Average Math Score"].map("{:.2f}".format)
districtSummary["Average Reading Score"] = districtSummary["Average Reading Score"].map("{:.2f}".format)
districtSummary["Percentage Passing Math"] = districtSummary["Percentage Passing Math"].map("{:.2f}%".format)
districtSummary["Percentage Passing Reading"] = districtSummary["Percentage Passing Reading"].map("{:.2f}%".format)
districtSummary["Percentage Overall Passing"] = districtSummary["Percentage Overall Passing"].map("{:.2f}%".format)


# Display the DataFrame
districtSummary
```

## Scores by School Spending

```python
#display the per_school_summary DF
per_school_summary.head()
```

```python
#display min and max of student budget
print(per_school_summary["Per Student Budget"].max())
print(per_school_summary["Per Student Budget"].min())
```

```python
# Establish the bins
spendingBins = [0, 580, 630, 650, 700]
labels = ["<$580", "$580-630", "$630-650", "$650<="]
```

```python
# Create a copy of the school summary for later aggregations
schoolSpendingDF = per_school_summary.copy()
schoolSpendingDF
```

```python
# Use `pd.cut` on the per_school_capita Series from earlier to categorize per student spending based on the bins.
schoolSpendingDF["Spending Ranges (Per Student)"] = pd.cut(perSchoolCapita, spendingBins, labels=labels)
schoolSpendingDF
```

```python
#converting columns back to floats
schoolSpendingDF["Average Math Score"] = schoolSpendingDF["Average Math Score"].astype(float)
schoolSpendingDF["Average Reading Score"] = schoolSpendingDF["Average Reading Score"].astype(float)
schoolSpendingDF["PCT Passing Math"] = schoolSpendingDF["PCT Passing Math"].str.replace("%","").astype(float)
schoolSpendingDF["PCT Passing Reading"] = schoolSpendingDF["PCT Passing Reading"].str.replace("%","").astype(float)
schoolSpendingDF["PCT Overall Passing"] = schoolSpendingDF["PCT Overall Passing"].str.replace("%","").astype(float)
#schoolSpendingDF
```

```python
#Check the Data Type
print(schoolSpendingDF.dtypes)
```

```python
#  Calculate averages for the desired columns.
spendingMathSeries = schoolSpendingDF.groupby(["Spending Ranges (Per Student)"])["Average Math Score"].mean()
spendingReadingSeries = schoolSpendingDF.groupby(["Spending Ranges (Per Student)"])["Average Reading Score"].mean()
spendingPassMathSeries = schoolSpendingDF.groupby(["Spending Ranges (Per Student)"])["PCT Passing Math"].mean()
spendingPassReadingSeries = schoolSpendingDF.groupby(["Spending Ranges (Per Student)"])["PCT Passing Reading"].mean()
spendingPassingSeries = schoolSpendingDF.groupby(["Spending Ranges (Per Student)"])["PCT Overall Passing"].mean()
```

```python
# Assemble into DataFrame
spendingSummaryDF = pd.DataFrame(
    {
        "Average Math Score": spendingMathSeries,
        "Average Reading Score": spendingReadingSeries,
        "PCT Passing Math": spendingPassMathSeries,
        "PCT Passing Reading": spendingPassReadingSeries,
        "PCT Overall Passing": spendingPassingSeries
    }
)

spendingSummaryDF["Average Math Score"] = spendingSummaryDF["Average Math Score"].map("{:.2f}".format)
spendingSummaryDF["Average Reading Score"] = spendingSummaryDF["Average Reading Score"].map("{:.2f}".format)
spendingSummaryDF["PCT Passing Math"] = spendingSummaryDF["PCT Passing Math"].map("{:.2f}%".format)
spendingSummaryDF["PCT Passing Reading"] = spendingSummaryDF["PCT Passing Reading"].map("{:.2f}%".format)
spendingSummaryDF["PCT Overall Passing"] = spendingSummaryDF["PCT Overall Passing"].map("{:.2f}%".format)
        
# Display results
spendingSummaryDF
```
